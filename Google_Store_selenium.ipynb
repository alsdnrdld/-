```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time, os
from datetime import datetime
import pandas as pd
```

```python


# review link link
link = 'https://play.google.com/store/apps/details?id=com.netflix.mediaclient&hl=ko&gl=US&showAllReviews=true'

# how many scrolls we need
scroll_cnt = 150

# download chrome driver https://sites.google.com/a/chromium.org/chromedriver/home
driver = webdriver.Chrome()
driver.implicitly_wait(6) 
driver.get(link)

os.makedirs('result', exist_ok=True)

for i in range(scroll_cnt):

    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    #time.sleep(3)
    try:
        load_more = driver.find_element_by_xpath('//*[contains(@class,"U26fgb O0WRkf oG5Srb C0oVfc n9lfJ")]').click()
        #time.sleep(3)
    except:
        print('더보기 버튼이 없네요~',i)

# 리뷰가 담긴 div영역에 접근하기
reviews = driver.find_elements_by_xpath('//*[@jsname="fk8dgd"]//div[@class="d15Mdf bAhLNe"]')

print('There are %d reviews avaliable!' % len(reviews))
print('Writing the data...')

# 데이터 담을 빈 데이터 프레임 만들기
df = pd.DataFrame(columns=['name', 'ratings', 'date', 'helpful', 'comment', 'manager_comment'])

# 리뷰 데이터 긁어오기
for review in reviews:
    # parse string to html using bs4
    resp = review.get_attribute('innerHTML')
    soup = BeautifulSoup(resp , 'html.parser')

    # 고객
    name = soup.find(class_='X43Kjb').text

    # rating (글자들 전부 지워버리고 별점 갯수 부분만 남긴 다음 숫자로 저장)
    ratings = int(soup.find('div', role='img').get('aria-label').replace('별표 5개 만점에', '').replace('개를 받았습니다.', '').strip())

    # review date
    date = soup.find(class_='p2TkOb').text
    date = datetime.strptime(date, '%Y년 %m월 %d일')
    date = date.strftime('%Y-%m-%d')

    # 좋아요
    helpful = soup.find(class_='jUL89d y92BAb').text
    if not helpful:
        helpful = 0 # 좋아요를 아에 누르지 않은 경우 null이 아닌 0으로 만들기
  
    # review text
    comment = soup.find('span', jsname='fbQN7e').text
    if not comment:
        comment = soup.find('span', jsname='bN97Pc').text
  
    # manager comment(담당자의 댓글)
    manager_comment = None
    dc_div = soup.find('div', class_='LVQB0b')
    if dc_div:
        manager_comment = dc_div.text.replace('\n', ' ')
  
  # 데이터 프레임에 저장하기
    df = df.append({
        'name': name,
        'ratings': ratings,
        'date': date,
        'helpful': helpful,
        'comment': comment,
        'manager_comment': manager_comment
      }, ignore_index=True)

# finally save the dataframe into csv file
filename = datetime.now().strftime('./result/NetFlex_%Y-%m-%d_%H-%M-%S.csv')
df.to_csv(filename, encoding='utf-8-sig', index=False)
driver.stop_client()
driver.close()

print('저장 완료!')
```
